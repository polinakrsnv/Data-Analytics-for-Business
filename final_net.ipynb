```python
from IPython.display import display, clear_output

from tqdm.notebook import tqdm
from sklearn.model_selection import train_test_split

import random

import torch
from torch import nn
from torch.nn import functional as F
from torch.utils.data import Dataset, DataLoader

import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

sns.set(style="darkgrid")
```

```python
SEED = 70
LR = 3e-2
WEIGHT_DECAY = 1e-3
NUM_EPOCHS = 15
GAMMA = 0.9995
BATCH_SIZE = 250
EVAL_BATCH_SIZE = 250
DEVICE = 'cpu'
```

```python
train_X = X_train
test_X = X_test
val_X = X_val
train_y = y_train
test_y = y_test
val_y = y_val
```

```python
class MyDataset(Dataset):
    def __init__(self, X, y):
        super().__init__()
        self.X = X
        self.y = y
        self.len = len(X)
        
    def __len__(self):
        return self.len
    
    def __getitem__(self, index):
        return self.X[index], self.y[index]
```

```python
train_dl = DataLoader(
    MyDataset(X_train, y_train),
    batch_size = BATCH_SIZE,
    shuffle = True
)

val_dl = DataLoader(
    MyDataset(X_val, y_val),
    batch_size = EVAL_BATCH_SIZE,
    shuffle = False
)

test_dl = DataLoader(
    MyDataset(X_test, y_test),
    batch_size = EVAL_BATCH_SIZE,
    shuffle = False
)

dls = {'train': train_dl, 'val': val_dl, 'test': test_dl}
```

```python
class Model(nn.Module):
    def __init__(self, in_features = 43, out_features = 1):
        super().__init__()
        self.in_features = in_features
        self.out_features = out_features
        self.hidden_size = int(np.sqrt(in_features + out_features))
        
        self. sequential = nn.Sequential(
            nn.Linear(in_features, self.hidden_size),
            nn.Sigmoid(),
            nn.Linear(self.hidden_size, 800),
            nn.Sigmoid(),
            nn.Linear(800, self.out_features)
        )
        
    def forward(self, x):
        x = self.sequential(x)
        return x
```

```python
train_set = MyDataset(X_train, y_train)
train_loader = train_dl

test_set = MyDataset(X_test, y_test)
test_loader = test_dl

model = Model()
#model.to(DEVICE) #DEVICE = 'cuda'

optimizer = torch.optim.AdamW(model.parameters(), lr=LR, weight_decay=WEIGHT_DECAY)
scheduler = torch.optim.lr_scheduler.ExponentialLR(optimizer, gamma=GAMMA)
criterion = nn.MSELoss()
```

```python
def train(model, optimizer, criterion, train_loader, test_loader):
    '''
    params:
        model - torch.nn.Module to be fitted
        optimizer - model optimizer
        criterion - loss function from torch.nn
        train_loader - torch.utils.data.Dataloader with train set
        test_loader - torch.utils.data.Dataloader with test set
                      (if you wish to validate during training)
    '''
    
    metrics_dict = {
        "Epoch": [],
        "Train MAE": [],
        "Val MAE": [],
        "Train MSE": [],
        "Val MSE": [],
        "Train RMSE": [],
        "Val RMSE": [],
    }

    # Train loop
    for epoch in tqdm(range(NUM_EPOCHS)):
        metrics_dict["Epoch"].append(epoch)
        for stage in ['train', 'val']:
            with torch.set_grad_enabled(stage == 'train'): # Whether to start building a graph for a backward pass
                if stage == 'train':
                    model.train() # Enable some "special" layers
                else:
                    model.eval() # Disable some "special" layers

                mae_loss_at_stage = 0
                mse_loss_at_stage = 0
                rmse_loss_at_stage = 0
                for batch in dls[stage]:
                    x_batch, y_batch = batch
                    x_batch, y_batch = x_batch.to(DEVICE), y_batch.to(DEVICE)

                    y_pred = model(x_batch.float()).view(-1) # forward pass: model(x_batch) -> calls forward()
                    loss = criterion(y_pred, y_batch.float()) # ¡Important! y_pred is always the first arg
                    if stage == "train":
                        loss.backward() # Calculate the gradients of all the parameters wrt loss
                        optimizer.step() # Update the parameters
                        scheduler.step()
                        optimizer.zero_grad() # Zero the saved gradient
                        
                    with torch.no_grad():
                        # Don't forget to rescale the results back to measure loss in original space
                        y_pred__ = y_pred * ss2.scale_ + ss2.mean_
                        y_batch__ = y_batch * ss2.scale_ + ss2.mean_
                        
                        mae_loss_at_stage += (torch.abs((y_pred__ - y_batch__)).sum()).item()
                        mse_loss_at_stage += (torch.square((y_pred__ - y_batch__)).sum()).item()
                        rmse_loss_at_stage += (torch.square((y_pred__ - y_batch__)).sum()).item()
                        
                mae_at_stage = (mae_loss_at_stage / len(dls[stage].dataset))
                metrics_dict[f"{stage.title()} MAE"].append(mae_at_stage)
                mse_at_stage = (mse_loss_at_stage / len(dls[stage].dataset))
                metrics_dict[f"{stage.title()} MSE"].append(mse_at_stage)
                rmse_at_stage = (rmse_loss_at_stage / len(dls[stage].dataset)) ** (1/2)
                metrics_dict[f"{stage.title()} RMSE"].append(rmse_at_stage)

        clear_output(wait=True)
        display(pd.DataFrame(metrics_dict))
        
        print(f'Epoch: {epoch + 1}')
    
        # Нарисуем график
        x = np.arange(len(metrics_dict['Train RMSE']))
        num_plots = 3
        total_cols = 3
        total_rows = 1
        
        fig, axs = plt.subplots(nrows=total_rows, ncols=total_cols,
                        figsize=(10*total_cols, 10*total_rows), constrained_layout=True)
        axs[0].set_title('Зависимость MAE', {'fontsize': 30})
        axs[0].set_xlabel('EPOCH', {'fontsize': 20})
        axs[0].set_ylabel('MAE', {'fontsize': 20})
        axs[0].tick_params(axis = 'both',    #  Применяем параметры к обеим осям
                   labelsize = 20)   #  Размер подписи
        axs[0].plot(x, metrics_dict['Train MAE'], color='blue', label='train mae')
        axs[0].plot(x, metrics_dict['Val MAE'], color='red', label='val mae')
        axs[0].legend(loc="upper right", prop={'size': 20})
        
        axs[1].set_title('Зависимость MSE', {'fontsize': 30})
        axs[1].set_xlabel('EPOCH', {'fontsize': 20})
        axs[1].set_ylabel('MSE', {'fontsize': 20})
        axs[1].tick_params(axis = 'both',    #  Применяем параметры к обеим осям
                   labelsize = 20)   #  Размер подписи
        axs[1].plot(x, metrics_dict['Train MSE'], color='blue', label='train mse')
        axs[1].plot(x, metrics_dict['Val MSE'], color='red', label='val mse')
        axs[1].legend(loc="upper right", prop={'size': 20})
        
        axs[2].set_title('Зависимость RMSE', {'fontsize': 30})
        axs[2].set_xlabel('EPOCH', {'fontsize': 20})
        axs[2].set_ylabel('RMSE', {'fontsize': 20})
        axs[2].tick_params(axis = 'both',    #  Применяем параметры к обеим осям
                   labelsize = 20)   #  Размер подписи
        axs[2].plot(x, metrics_dict['Train RMSE'], color='blue', label='train rmse')
        axs[2].plot(x, metrics_dict['Val RMSE'], color='red', label='val rmse')
        axs[2].legend(loc="upper right", prop={'size': 20})

        plt.show()
```

```python
def mae(preds, y_true):
    # Don't forget to rescale the results back to measure loss in original space
    preds__ = preds * ss2.scale_ + ss2.mean_
    y_true__ = y_true * ss2.scale_ + ss2.mean_
    return mean_absolute_error(y_true__, preds__)
```

```python
def mse(preds, y_true):
    # Don't forget to rescale the results back to measure loss in original space
    preds__ = preds * ss2.scale_ + ss2.mean_
    y_true__ = y_true * ss2.scale_ + ss2.mean_
    return mean_squared_error(y_true__, preds__)
```

```python
def rmse(preds, y_true):
    # Don't forget to rescale the results back to measure loss in original space
    preds__ = preds * ss2.scale_ + ss2.mean_
    y_true__ = y_true * ss2.scale_ + ss2.mean_
    return mean_squared_error(y_true__, preds__, squared = False)
```

```python
def eval_loop(model, X_val, X_test, y_val, y_test, model_is_nn=True, model_name="NN"):
    print(f"Model {model_name}")
    for name, x_eval, y_eval in zip(['val', 'test'], [X_val, X_test], [y_val, y_test]):
        if model_is_nn:
            with torch.no_grad():
                preds = model(torch.Tensor(x_eval)).view(-1)
        else:
            model.fit(X_train, y_train)
            preds = model.predict(x_eval)
        mae_val = mae(preds, y_eval)
        mse_val = mse(preds, y_eval)
        rmse_val = rmse(preds, y_eval)
        print(f'{name}: MAE = {mae_val:.2f}')
        print(f'{name}: MSE = {mse_val:.2f}')
        print(f'{name}: RMSE = {rmse_val:.2f}')
```

```python
def test(model, criterion, test_loader):
    '''
    params:
        model - torch.nn.Module to be evaluated on test set
        criterion - loss function from torch.nn
        test_loader - torch.utils.data.Dataloader with test set
    ----------
    returns:
        predicts - torch.tensor with shape (len(test_loader.dataset), ),
                   which contains predictions for test objects
    '''

    name, x_eval, y_eval = 'test', X_test, y_test
    model.eval() # Disable some "special" layers
    with torch.no_grad():
        preds = model(torch.Tensor(x_eval)).view(-1)

    print("Test MAE = %.2f" % mae(preds, y_eval))
    print("Test MSE = %.2f" % mse(preds, y_eval))
    print("Test RMSE = %.2f" % rmse(preds, y_eval))
        
    return preds
```
